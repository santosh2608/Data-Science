# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func, inspect
from datetime import datetime, timedelta
```

```python
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)
```

```python
# We can view all of the classes that automap found
Base.classes.keys()
```

```python
# get column details for measurement and station table 

inspector = inspect(engine)

cols = inspector.get_columns('measurement')
print("Measurement\n------------")
for col in cols:
    print(col['name'], col['type'])
    
cols = inspector.get_columns('station')
print("\n\n Station\n-------------")
for col in cols:
    print(col['name'], col['type'])
```

```python
# Save references to each table
Measurements = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

```python
# Grabs the last date entry in the data table
last_date = session.query(Measurements.date).order_by(Measurements.date.desc()).first()
print(last_date)
```

```python
# to get the last 12 months of data, last date - 365
day = 365

# to get the last 12 months of data, last date - 365
last_year = dt.date(2017, 8, 23) - dt.timedelta(days=day)
print(last_year)
```

# Exploratory Climate Analysis

```python
# Query to retrieve the last 12 months of precipitation data and plot the results
# Calculate the date 1 year ago from the last data point in the database

rain_prcp = session.query(Measurements.date, Measurements.prcp).\
    filter(Measurements.date > last_year).order_by(Measurements.date).all()
```

```python
# Save the query results as a Pandas DataFrame and set the index to the date column

rain_prcp_df = pd.DataFrame(rain_prcp)

# reset the index to the date
rain_prcp_df.set_index('date').head()

rain_prcp_df.head()
```

```python
# Use Pandas Plotting with Matplotlib to plot the data
rain_prcp_df.plot('date', 'prcp', figsize=(15,10), color="purple")
# Set the x axis label
plt.xlabel("Date", fontsize=12)
# Set the y axis label
plt.ylabel("Rain in Inches", fontsize=12)
# Set the chart title
plt.title("Last 12 months of precipitation", fontsize=14)
# Set the legend
plt.legend(["Precipitation"])
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("Result/12_months_precipitation.png")
plt.show()
```

```python
# Design a query to show how many stations are available in this dataset?
stations = session.query(Measurements).group_by(Measurements.station).count()
print("There are {} stations available in the dataset.".format(stations))
```

```python
# What are the most active stations? (i.e. what stations have the most rows)?
# List the stations and the counts in descending order.
busy_station = session.query(Measurements.station, func.count(Measurements.tobs)).group_by(Measurements.station).\
               order_by(func.count(Measurements.tobs).desc()).all()

busiest = busy_station[0][0]    
print("The most active Station was",busiest,"with",busy_station[0][1],"weather observations.\n")

busy_station
```

```python

# Using the station id from the previous query, calculate the lowest temperature recorded, 
# highest temperature recorded, and average temperature most active station?

temp_analysis = session.query(func.min(Measurements.tobs), func.max(Measurements.tobs), func.avg(Measurements.tobs)).\
    filter(Measurements.station == busiest).\
    filter(Measurements.date > last_year).\
    order_by(Measurements.date).all()

print (temp_analysis)
```

```python
# Using the station id from the previous query, calculate the lowest temperature recorded, 
# highest temperature recorded, and average temperature most active station?
temperature = session.query(Measurements.station, Measurements.date, Measurements.tobs).\
    filter(Measurements.station == busiest).\
    filter(Measurements.date > last_year).\
    order_by(Measurements.date).all()
temperature
```

```python
# Choose the station with the highest number of temperature observations.
# Query the last 12 months of temperature observation data for this station and plot the results as a histogram
# Create dataframe with temprature data
temperature_df=pd.DataFrame(temperature)
# Plot histogram
plt.hist(temperature_df['tobs'], 12)
# Set x axis label
plt.xlabel("Recorded Temperature")
# Set y axis label
plt.ylabel("Frequency")
# Set title
plt.title("Busiest Station " + busiest + " Analysis \n")
plt.savefig("Result/temp_frequency.png")
plt.legend(["tobs"])
plt.show()
```

```python
# This function called `calc_temps` will accept start date and end date in the format '%Y-%m-%d' 
# and return the minimum, average, and maximum temperatures for that range of dates
def calc_temps(start_date, end_date):
    """TMIN, TAVG, and TMAX for a list of dates.
    
    Args:
        start_date (string): A date string in the format %Y-%m-%d
        end_date (string): A date string in the format %Y-%m-%d
        
    Returns:
        TMIN, TAVE, and TMAX
    """
    
    return session.query(func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)).\
        filter(Measurement.date >= start_date).filter(Measurement.date <= end_date).all()

# function usage example
print(calc_temps('2012-02-28', '2012-03-05'))
```

```python
# Use your previous function `calc_temps` to calculate the tmin, tavg, and tmax 
# for your trip using the previous year's data for those same dates.

trip_arrive = dt.date(2018, 8, 20)
trip_leave = dt.date(2018, 8, 26)
last_year = dt.timedelta(days=day)
avg_temp_lst_year = (calc_temps((trip_arrive-last_year), (trip_leave-last_year)))

print ("Temprature on my summar vacation between " + str(trip_arrive-last_year) + " to " + str(trip_leave-last_year))
print("\n")
print(f"Lowest Temprature : {avg_temp_lst_year[0][0]} F")
print(f"Highest Temprature : {avg_temp_lst_year[0][2]} F")
print("Average Temprature : {:.1f} F".format(avg_temp_lst_year[0][1]))

#print(calc_temps('2012-02-28', '2012-03-05'))
```

```python
# Use the average temperature for the y value
# Use the peak-to-peak (tmax-tmin) value as the y error bar (yerr)

print(f"Lowest Temprature : {avg_temp_lst_year[0][0]} F")
print(f"Highest Temprature : {avg_temp_lst_year[0][2]} F")
print("Average Temprature : {:.1f} F".format(avg_temp_lst_year[0][1]))

# Plot the results from your previous query as a bar chart. 
fig, ax = plt.subplots()
x = range(len(avg_temp_lst_year))
ax.boxplot(avg_temp_lst_year, patch_artist=True)
# Use "Trip Avg Temp" as your Title
ax.set_title('Trip Avg Temp \n')
# Set y axis label
ax.set_ylabel("Temp (F)")
fig.tight_layout()
plt.savefig("Result/trip_avg_temp.png")
plt.show()

```

```python
# Calculate the total amount of rainfall per weather station for your trip dates using the previous year's matching dates.
# Sort this in descending order by precipitation amount and list the station, name, latitude, longitude, and elevation

print ("\n")
print ("Total amount of rainfall per weather station for my trip between "+ str(trip_arrive-last_year)+ " to " + str(trip_leave-last_year))
print ("\n")
total_amt_of_rainfall = session.query(Measurements.station, Station.name, Station.latitude, Station.longitude, Station.elevation, Measurements.date).\
filter(Measurements.date >= (trip_arrive-last_year)).\
    filter(Measurements.date <= (trip_leave-last_year)).\
    order_by(Measurements.station).all()

#print (total_amt_of_rainfall)
total_amt_of_rainfall_df = pd.DataFrame(total_amt_of_rainfall)
total_amt_of_rainfall_df
```

```python
# Create a query that will calculate the daily normals 
# (i.e. the averages for tmin, tmax, and tavg for all historic data matching a specific month and day)

def daily_normals(date):
    """Daily Normals.
    
    Args:
        date (str): A date string in the format '%m-%d'
        
    Returns:
        A list of tuples containing the daily normals, tmin, tavg, and tmax
    
    """
    
    sel = [func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)]
    return session.query(*sel).filter(func.strftime("%m-%d", Measurement.date) == date).all()
    
daily_normals("01-01")
```

```python
# calculate the daily normals for your trip
# push each tuple of calculations into a list called `normals`

# Set the start and end date of the trip

# Use the start and end date to create a range of dates

# Stip off the year and save a list of %m-%d strings

# Loop through the list of %m-%d strings and calculate the normals for each date


# str(trip_arrive-last_year)+ " to " + str(trip_leave-last_year)
arrive_date = trip_arrive-last_year
leave_date = trip_leave-last_year
vacation_days = 6

print(f"Arrival Day   : {arrive_date}")
print(f"Departure Day : {leave_date}")
print (f"Vacation Duration: {vacation_days}")

dates = [arrive_date + dt.timedelta(days = i) for i in range(vacation_days)]
# Loop through the list of %m-%d strings and calculate the normals for each date
normal = []
for i in range(vacation_days):
    day_normal = daily_normals(func.strftime("%m-%d", arrive_date + dt.timedelta(days = i)))
    normal.append(day_normal[0])
normal

```

```python
# Load the previous query results into a Pandas DataFrame and add the `trip_dates` range as the `date` index
normal_df = pd.DataFrame(normal ,columns=['tmin','tavg','tmax'])
# Add date to the dataframe
normal_df['date'] = dates
# range as the date index
normal_df.set_index('date', inplace = True)
normal_df
```

```python
# Plot the daily normals as an area plot with `stacked=False`
x = np.arange(vacation_days)
# Set figure size
plt.figure(figsize=(15, 8))
# set vorder 
plt.plot(x, normal_df['tmin'], color= 'sienna', label = "tmin")
plt.plot(x, normal_df['tavg'], color= 'coral', label = "tavg")
plt.plot(x, normal_df['tmax'], color= 'goldenrod', label = "tmax")
# fill colour to the respective parameter
plt.fill_between(x, normal_df['tmin'], color = 'sienna', alpha = 0.5)
plt.fill_between(x, normal_df['tavg'],normal_df['tmin'], color = 'coral', alpha = 0.5)
plt.fill_between(x, normal_df['tmax'],normal_df['tavg'], color = 'goldenrod', alpha = 0.5)
# set x label as vacation days
plt.xticks(x, dates, rotation = 45)
# set x label
plt.xlabel("\nDates", fontsize = 12)
# set y label
plt.ylabel('Temp (F)\n', fontsize = 12)
# set figure title
plt.title("Temperature Prediction on vacation days \n", fontsize = 20)
# set legend font size
plt.legend(fontsize = 15)
plt.xlim(x.min(), x.max())
plt.ylim(0,normal_df['tmax'].max() + 5)
plt.tight_layout()
# save figure to the result folder
plt.savefig("Result/daily_tempe_during_vacation.png")
plt.show()
```
